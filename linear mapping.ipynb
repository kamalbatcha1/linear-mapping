```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
v=np.array([-1,2])
i_hat=np.array([1,0])
j_hat=np.array([0,1])
v_ihat=v[0]*i_hat
v_jhat=v[1]*j_hat
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v_ihat,color="g",linestyle="dotted",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(v_ihat[0],v_ihat[1],*v_jhat,color="r",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(0,0,*v,color="b",linewidth=2.5,head_width=.30,head_length=.35)
plt.xlim(-4, 2)
major_xticks = np.arange(-4, 2)
ax.set_xticks(major_xticks)
plt.ylim(-2, 4)
major_yticks = np.arange(-2, 4)
ax.set_yticks(major_yticks)
plt.grid(b=True, which='major')
plt.show()
```

```python
# Define vector v 
v = np.array([-1, 2])
ihat_t=np.array([3,1])
jhat_t=np.array([1,2])
v_ihat_t = v[0] * ihat_t
v_jhat_t = v[1] * jhat_t
v_t=v_ihat_t+v_jhat_t
ax = plt.axes()
ax.plot(0,0,'or')
ax.arrow(0, 0, *v_ihat_t, color='g', linestyle='dotted', linewidth=2.5, head_width=0.30,head_length=0.35)
ax.arrow(v_ihat_t[0], v_ihat_t[1], *v_jhat_t, color='r', linestyle='dotted', linewidth=2.5,head_width=0.30, head_length=0.35)
ax.arrow(0, 0, *v, color='skyblue', linewidth=2.5, head_width=0.30, head_length=0.35)
ax.arrow(0,0,*v_t,color="b", linewidth=2.5, head_width=0.30, head_length=0.35)
plt.xlim(-4, 2)
major_xticks = np.arange(-4, 2)
ax.set_xticks(major_xticks)
plt.ylim(-2, 4)
major_yticks = np.arange(-2, 4)
ax.set_yticks(major_yticks)
plt.grid(b=True, which='major')
plt.show()
```
