```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
v=np.array([1,1])
ax=plt.axes()
```

```python
ax=plt.axes()
ax.plot(0,0,"or")
```

```python
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v,color="b",linewidth=2.0,head_width=.20,head_length=.25)
plt.xlim(-2,2)
major_xticks=np.arange(-2,3)
ax.set_xticks(major_xticks)
plt.ylim(-1,2)
major_yticks=np.arange(-1,3)
ax.set_yticks(major_yticks);
plt.grid(b=True,which="major")
plt.show()
```

```python
v=np.array([1,1])
a=3
av=a*v
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v,color="b",linewidth=2.0,head_width=.20,head_length=.25)
ax.arrow(0,0,*av,color="c",linestyle="dotted",linewidth=2.0,head_width=.20,head_length=.25)
plt.xlim(-2,4)
major_xticks=np.arange(-2,4)
ax.set_xticks(major_xticks)
plt.ylim(-1,4)
major_yticks=np.arange(-1,4)
ax.set_yticks(major_yticks)
plt.grid(b=True,which="major")
plt.show()
```

```python
v=np.array([1,1])
w=np.array([-2,2])
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v,color="b",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(0,0,*w,color="c",linewidth=2.5,head_width=.30,head_length=.35)
plt.xlim(-3,2)
major_xticks=np.arange(-3,2)
ax.set_xticks(major_xticks)
plt.ylim(-1,4)
major_yticks=np.arange(-1,4)
ax.set_yticks(major_yticks)
plt.grid(b=True,which="major")
plt.show()
```

```python
v=np.array([1,1])
w=np.array([-2,2])
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v,color="b",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(v[0], v[1], *w, linestyle='dotted',color="c",linewidth=2.5,head_width=.30,head_length=.35)
plt.xlim(-3,2)
major_xticks=np.arange(-3,2)
ax.set_xticks(major_xticks)
plt.ylim(-1,4)
major_yticks=np.arange(-1,4)
ax.set_yticks(major_yticks) 
plt.grid(b=True,which="major")
plt.show()
```

```python
v=np.array([1,1])
w=np.array([-2,2])
vw=v+w
ax=plt.axes()
ax.plot(0,0,"or")
ax.arrow(0,0,*v,color="b",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(v[0], v[1], *w, linestyle='dotted',color="c",linewidth=2.5,head_width=.30,head_length=.35)
ax.arrow(0,0,*vw,color="k",linewidth=2.5,head_width=.30,head_length=.35)
plt.xlim(-3,2)
major_xticks=np.arange(-3,2)
ax.set_xticks(major_xticks)
plt.ylim(-1,4)
major_yticks=np.arange(-1,4)
ax.set_yticks(major_yticks)
plt.grid(b=True,which="major")
plt.show()
```
